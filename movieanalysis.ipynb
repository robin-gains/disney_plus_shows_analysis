*Weighted formula to calculate meta scores : Missing At Random*

```python
def new_metas(ratings, votes):
    return ((votes / (6000 + votes)) * ratings) + ((6000 / (votes + 6000)) * 7.0)
```

```python
dataframe['meta'] = new_metas(dataframe['imdb_rating'],dataframe['imdb_votes'])
```

```python
dataframe['meta'] = dataframe['meta'].round(3)
```

```python
dataframe['meta'] = dataframe['meta'] * 10
```

```python
missing_data = {
                'column' : ['genre', 'director','writer','awards'],
                'Missing_percentage' : [0.1, 19.9, 13.8, 33.3]
}
```

```python
missing_values = pd.DataFrame(missing_data)
```

*PLOT TO FIND MISSING PERCENTAGE*

```python
plt.figure(figsize=(10,6))
ax = sns.barplot(x='column', y='Missing_percentage', data=missing_values)

# Loop through bars and annotate
for p in ax.patches:
    ax.text(
        p.get_x() + p.get_width()/2,   # X position (center of bar)
        p.get_height(),                # Y position (top of bar)
        f'{p.get_height():.2f}%',      # Text (formatted percentage)
        ha='center', va='bottom'       # Align center horizontally, bottom vertically
    )

plt.title('Percentage of missing values')
plt.ylabel('Percentage(%)')
plt.xlabel('Columns')
plt.ylim(0, 100)
plt.show()
```
